```python
#Importing Required Libraries

import pandas as pd
```

```python
#Loading Data

transactions_df = pd.read_csv("transactions_train_subset.csv")
customers_df = pd.read_csv("customers_subset.csv")
articles_df = pd.read_csv("articles_subset.csv")
```

## Construct New Features 

```python
# Contructing Global Purchase Pattern Features
# We perform group by over customer_id to calculate the 
# max,min,median and mean purchase value of each customer
new_features_df = pd.DataFrame(transactions_df.groupby(['customer_id']).aggregate(['min','max','mean','median'])['price'])
```

```python
# We rename the aggregate columns for the purpose of better naming convention
new_features_df = new_features_df.rename(columns={"mean": "Avg_Price_Transaction",
                                "min": "Min_Price_Transaction",
                                "max": "Max_Price_Transaction",
                                "median": "Median_Price_Transaction"}).head()
```

```python
new_features_df = new_features_df.reset_index()
new_features_df.head()
```

```python
# Construction of Group Level Purchase Patterns
# We join the articles dataframe with transactions dataframe 
# to take into the articles attributes into account as well
combined_df = transactions_df.join(articles_df, how='inner',lsuffix='article_id', rsuffix='article_id').reset_index()
```

```python
combined_df.head()
```

```python
# We perform group by over customer_id,section_name,department_name,garment_group_name to calculate the 
# max,min,median and mean purchase value of each customer at the section and garment group level in each department
# This has been done as the amount we spend for each type of clothing differs from person to person

df = combined_df.copy()

x = df.groupby(['customer_id','section_name','department_name','garment_group_name']
            ).agg({'price':["mean","min","max","median"]})
```

### Merging All New features

```python
# Convert Index into Columns

x = x.reset_index()

#Rename columns for better naming Convention
x = x.rename(columns={"mean": "Group_Level_mean_Price_Transaction",
                                "min": "Group_Level_Min_Transaction",
                                "max": "Group_Level_Max_Transaction",
                                "median": "Group_Level_Median_Transaction"}).head()

# Then Convert Multilevel Indexed columns to Single level columns
x.columns = x.columns.map(''.join)
```

```python

x.head()
```

```python
# Perform Join operation to have one df containing all new features
new_features = new_features_df.join(pd.DataFrame(x), how='inner',lsuffix='', rsuffix='customer_id')
new_features = new_features.drop(['customer_idcustomer_id'],axis=1)
```
